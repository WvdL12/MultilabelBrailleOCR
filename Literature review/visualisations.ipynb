### Figures

```python
size_map = {
    1: 3,
    2: 3,
    3: 5,
    4: 5,
    5: 5,
    6: 5,
    7: 6,
    8: 6,
    9: 6,
    10: 7,    
}
```

```python
filter_list = [
 'Image Acquisition',
 'One-sided vs Inter-point',
 'Pre-processing',
 'Dot Detection',
 'Character Recognition',
 'Translation',
 'Hybrid Systems',
 'CNN Systems',
 'Languages',
 'Datasets'
]
# jitters = [-0.15, -0.1, -0.05, 0, 0.05, 0.1, 0.15]
JS = 0.175
jitters = [0, JS, -JS]
max_cite = 30

norm = plt.Normalize(0, max_cite)
sm = plt.cm.ScalarMappable(cmap=CONPAL_CMAP, norm=norm)
for d, (df, df_name, df_abb) in enumerate(zip(dfs, df_names, df_abbrev)):
    if df_name not in filter_list:
        continue
    S = size_map[df_sizes[d]]
    Y = df.columns[0]
    fig, axes = plt.subplots(1, 1, figsize=(9, max(4, S)))
    # fig, axes = plt.subplots(1, 1, figsize=(9, max(4, np.ceil(S / 2) + 2)))

    # df['Year Jitter'] = (np.random.choice(jitters, size=df.shape[0], replace=True)) + df['Year']
    # Group by Technique and Year, then assign jitters cyclically within each group
    df['Year Jitter'] = df.groupby([Y, 'Year']).cumcount() % len(jitters)
    df['Year Jitter'] = df['Year Jitter'].map(lambda x: jitters[int(x)]) + df['Year']
    
    plot_ax = sns.swarmplot(data=df, x='Year Jitter', y=Y, ax=axes, hue='Citations', palette=CONPAL_CMAP, dodge=False, size=7, zorder=1, legend=False)
    # plot_ax.get_legend().remove()
    cbar = plot_ax.figure.colorbar(sm, ax=plot_ax, orientation='horizontal', 
                                fraction=0.03, pad=.9 / S, aspect=50)
    # plot_ax.figure.colorbar(sm, ax=plot_ax, label='Number of Citations', orientation='horizontal')
    cbar.ax.tick_params(labelsize=TICK_SIZE)
    cbar.ax.set_xlabel('Times cited', fontsize=LBL_SIZE)
    axes2 = axes.twiny()
    bars = sns.histplot(data=df, y=Y, color=CPAL[2], alpha=0.5, fill=False, linewidth=2, shrink=0.5, ax=axes2)
    ylim = axes.get_ylim()
    ylim_adj = ylim[0]*0.05
    axes.set_ylim(ylim[0]+ylim_adj, ylim[1]-ylim_adj)
    xlim = axes2.get_xlim()[1]*1.15
    axes2.set_xlim(0, xlim)
    for p in bars.patches:
        count = int(p.get_width())
        bars.annotate(f" Total: {count}", 
                    (xlim, p.get_y() + p.get_height() / 2), 
                    ha='left', va='center', zorder=-1, alpha=0.5, weight='bold', size=LBL_SIZE)
    # axes.set_title(f"Prevalence of different applications in {df_name}")
    # axes.set_xlim([1985, 2025])
    axes2.set_xticks([], [])
    axes2.set_xlabel('')
    axes2.tick_params(axis='x', which='both', bottom=False)
    axes.tick_params(axis='x', which='both', bottom=False)
    # axes2.set_ylabel('Total')
    # axes.legend(loc='center left', bbox_to_anchor=(1.05, 0.5), title='Number of citations')
    axes.xaxis.set_major_locator(MaxNLocator(integer=True))
    axes.set_xlabel('', fontsize=LBL_LARGE)
    axes.set_ylabel("")
    # axes.set_title(Y, fontsize=LBL_LARGE)
    axes.tick_params(axis='both', labelsize=TICK_SIZE)
    plt.tight_layout()
    plt.savefig(os.path.join('plots', f'lit_{df_abb}_{version}'), bbox_inches='tight')
```
